# Probabilità

L'incertezza può essere rappresentata come un certo numero di eventi e la probabilità che accadano.

## Parole possibili

Ogni possibile situazione può essere pensata come un "mondo", rapprentato dalla lettera $ω$

Ad esempio, un dado può risultare in 6 mondi differenti: uno dove il risultato del dado è 1, uno dove è 2, ecc...

Per rappresentare la probabilità di un certo evento $ω$ scriveremo
: $P(ω)$

## Assiomi della Probabilità

- $0 \leq P(ω) \leq 1$
  - ogni valore assunto dalla probabilità deve essere compreso tra 0 e 1
  - $0$ è un evento impossibile
  - 1 è un evento che accadrà sicuramente
  - In generale, più ci avviciniamo a 1 più è possibile che l'evento accada

- $La\ somma\ delle\ probabilità\ di\ ogni\ possibile\ evento\ è\ pari\ a\ 1$


### Probabilità incondizionata

## Probabilità Condizionata

Grado di fiducia in una proposizione date prove che sono già state rivelate.

Viene rappresentata come $P(a|b)$, e sta per: Probabilità che $a$ sia vera quando $b$ è vera.

Mi interessa sapere, quindi, la probabilità di $a$ vera **solo** quando $b$ è vera, ignorando i valori quando $b$ è falsa. La formula è:



**Esempi concettuali:**

- $P(pioggia\ oggi\ |\ piovuto\ ieri)$
- $P(cambio\ strada\ |\ condizioni\ traffico)$
- $P(malattia\ |\ risultato\ esame)$

**Esempio matematico:**

Dati 2d6, uno blu e uno rosso, quanto vale $P(somma\ 12|\ rosso=6)$?

$P(somma\ 12|\ rosso=6)=\frac{P(somma\ 12\ ∧\ rosso=6)}{P(rosso=6)}$

Tuttavia:

- $P(somma\ 12\ ∧\ rosso=6)=\frac{1}{36}$ poiché ottengo $12$ solo quando entrambi i dadi valgono 6
- $P(rosso=6)=\frac{1}{6}$ poiché il 6 è solo una delle sei facce del dado rosso

Ottengo quindi:

$P(somma\ 12|\ rosso=6)=\frac{\frac{1}{36}}{\frac{1}{6}}=\frac{1}{6}$

## Esempio:

Voglio capire quanto è probabile che piova nel pomeriggio dato il fatto che stamattina c'erano nuvole: $P(pioggia\ |\ nuvole)$.

Cominciamo con le seguenti informazioni:

$P(pioggia\ |\ nuvole)=\frac{P(pioggia)P(nuvole\ |\ pioggia)}{P(nuvole)}$

Conoscere $P(a \mid b)$, oltre a $P(a)$ e $P(b)$, ci permette di calcolare $P(b \mid a)$. Questo è utile, perché conoscere la probabilità condizionata di un effetto visibile data una causa sconosciuta, $P(effetto\ visibile\ \mid\ causa\ sconosciuta)$, ci permette di calcolare la probabilità della causa sconosciuta dato l’effetto visibile, $P(causa\ sconosciuta \mid\ effetto\ visibile)$.

Ad esempio, possiamo apprendere $P(risultati\ esame\ \mid\ malattia)$ attraverso studi clinici, testando persone con la malattia e osservando quanto spesso il test la rileva. Conoscendo questo, possiamo calcolare $P(malattia\ \mid\ risultati\ esame)$, che è un'informazione diagnostica preziosa.


# Probabilità congiunta

Probabilità che più eventi accadano contemporaneamente.

Ritorniamo all'esempio

| **C = cloud** | **C = ¬cloud** |
| --- | --- |
| 0.4 | 0.6 |

<br>

| **R = rain** | **R = ¬rain** |
| --- | --- |
| 0.1 | 0.9 |

Avendo solo queste informazioni, non so in che modo i due eventi possano essere collegati tra loro. Con la tabella delle probabilità congiunte, invece, possiamo ottenere altre informazioni. La tabella è la combinazione di tutti i valori sopra.

| | **R = rain** | **R = ¬rain** |
| ---| --- | --- |
| **C = cloud** | 0.08 | 0.32 |
| **C = ¬cloud** | 0.02 | 0.58|

A partire dalla tabella della probabilità congiunta, posso calcolare quale è la probabilità che la mattina sia nuvolosa vista la pioggia nel pomeriggio. Nota che con $C$, data la non-specificazione, rappresento 2 valori: cloud e ¬cloud

$P(C|\ R=rain)$

Sto quindi cercando una distribuzione: la risposta non sarà un solo valore, ma un vettore di due, la cui somma degli elementi è 1.

La probabilità sopracitata sarà quindi uguale alla probabilità che sia nuvoloso e che piova, diviso la probabilità che piova.

$P(C|\ R=rain)=\frac{P(C,R=rain)}{P(R=rain)}$

$\frac{1}{P(R=rain)}$, in fin dei conti, non è altro che l'inverso di una costante. Per ora la indichiamo con $α$, ci saranno spiegazioni più avanti. Otteniamo quindi:

$P(C|\ R=rain)=αP(C,R=rain)=α<0.08,\ 0.02>$

La probabilità condizionata, quindi, risulta proporzionale alla probabilità congiunta.

Normalizzo per far sì che le componenti del vettori si sommino a 1: $P(C|\ R=rain)=<0.8,\ 0.2>$

# Regole di probabilità

## Negazione

$P(¬a) = 1 - P(a)$
  - La somma di tutte le probabilità di ogni evento deve essere pari a 1, e ciò include sia $a$ che il suo negato

## Inclusione-Esclusione

$P(a ∨ b) = P(a) + P(b) - P(a ∧ b)$
  - La probabilità che $a$ o $b$ sia vero è uguale alla probabilità che $a$ sia vero più la probabilità che $b$ sia vero. Tuttavia, in questo modo alcune probabilità vengono contate due volte: nello specifico quando sia $a$ che $b$ sono vere. Sottraggo quindi la probabilità della loro and alla somma delle loro probabilità singole.
  - Esempio: Oggi mangio carne: probabilità 80% ; oggi mangio pesce: probabilità 70%. Se calcolo la probabilità che che oggi mangio carne o pesce senza sottrarre la probabilità di carne *e* pesce, ottengo una probabilità di 0.8+0.7=1.5, che vìola il primo assioma.

## Marginalizzazione

$P(a) = P(a, b) + P(a, ¬b)$
  - L'idea è che $b$ e $¬b$ sono probabilità indipendenti, cioè $P(b,¬b)=0$. Sappiamo anche che $P(b)+P(¬b)=1$. Quindi, quando si verifica $a$, $b$ può accadere o meno. Valutando, quindi, la probabilità che $a$ e $b$ accadano insieme e sommandola alla probabilità che $a$ e $¬b$ accadano insieme, ottengo semplicemente la probabilità di $a$

### Marginalizzazione per variabili aleatorie

La probabilità della $i$-esima variabile aleatoria è la somma delle $j$-esime probabilità congiunte.



Riprendendo l'esempio sopra, la probabilità che sia nuvoloso è dato dalla somma delle probabilità congiunte sia nel caso che piova sia nel caso che non piova:

$P(C = cloud) = P(C = cloud, R = rain) + P(C = cloud, R = ¬rain) = 0.08 + 0.32 = 0.4$

## Condizionamento:

$P(a) = P(a | b)P(b) + P(a | ¬b)P(¬b)$
  - Idea simile alla marginalizzazione

### Condizionamento per variabili aleatorie

La probabilità della $i$-esima variabile aleatoria è la somma delle $j$-esime probabilità congiunte moltiplicata per la probabilità individuale della $j$-esima variabile aleatoria.



# Rete Bayesiana

Struttura dati che rappresenta le dipendenze tra variabili aleatorie.

## Proprietà delle Reti Bayesiane

- Sono grafi direzionati
- Ogni nodo del grafo rappresenta una variabile aleatoria
- Una freccia da un nodo $X$ a un nodo $Y$. Rappresenta la distribuzione di probabilità che $Y$ dipende da $X$
- Ogni nodo $X$ ha una distribuzione di probabilità $P(X \mid Parents(X) )$

### Esempio di Rete Bayesiana

Sto cercando di arrivare in tempo ad un appuntamento.



#### Descrizione

Descriviamo la rete dall'alto verso il basso

1. **Rain**

Siamo alla root (radice) del grafo. La distribuzione probabilistica non dipende da nessun evento antecedente.

Nel nostro esempio, `Rain` è una variabile aleatoria che assume valori `{none, light, heavy}` con le seguenti distribuzioni:

| none | light | heavy |
| --- | --- | --- |
| 0.7 | 0.2 | 0.1 |

2. **Maintenance**

Nel nostro esempio, `Maintenance` codifica se c'è manutenzione della ferrovia, assumendo valori `{yes, no}`.

`Rain` è il nodo genitore di `Maintenance`, quindi la distribuzione di probabilità di `Maintenance` è influenzata da `Rain`

| Rain | yes | no |
| --- | --- | --- |
|none | 0.4 | 0.6 |
| light | 0.2 | 0.8 |
| heavy | 0.1 | 0.9 |

3. **Train**

Nel nostro esempio, `Train` codifica se il treno è in orario o in ritardo, assumendo i valori `{on time, delayed}`.

Nota che `Train` ha frecce che puntano ad esso sia da `Maintenance` che da `Rain`. Ciò significa che sono entrambi genitori di `Train` e i loro valori influenzano la distribuzione di probabilità di `Train`

| Rain | Maintenance | on time | delayed |
| --- | --- | --- | --- |
|none | yes | 0.8 | 0.2 |
|none | no | 0.9| 0.1 |
|light | yes | 0.6 | 0.4 |
|light | no | 0.7 | 0.3 |
|heavy | yes | 0.4 | 0.6 |
|heavy | no | 0.5 | 0.5 |

4. **Appointment**

Nell'esempio, `Appointment` è una variabile aleatoria che rappresenta se ci presentiamo all'appuntamento o meno, e assume valori di `{attend, miss}`.

Nota che l'unico genitore è `Train`. I genitori sono solo relazioni dirette. È vero che `Maintenance` ha effetto sul fatto che il treno sia in orario o meno, e il se il treno è in orario o no influisce sulle nostre possibilità di presentarci all'appuntamento. Alla fine, però, ciò che influenza le nostre probabilità di presentarci all'appuntamento è solo il fatto che il treno sia in orario o meno, che è ciò che viene rappresentato dalla Rete Bayesiana.

Ad esempio, se il treno è in orario, potrebbe esserci pioggia pesante e manutenzioni sulla linea, ma ciò non ha effetto sul fatto che ci siamo presentati o meno all'appuntamento.

| Train | attend | miss |
| --- | --- | --- |
| on time | 0.9 | 0.1 |
| delayed | 0.6 | 0.4 |

Se volessimo trovare la probabilità di mancare all'appuntamento quando il treno è in ritardo in un giorno senza manutenzione e pioggia leggera, cioè $P(light,\ no,\ delayed,\ miss)$, scriveremo la seguente formula:

$P(light)P(no \mid light)P(delayed \mid light, no)P(miss \mid delayed)$

Il valore di tutte queste probabilità è presente nelle tabelle sopra, e saranno tutte moltiplicate per ottenere il valore di: $P(light,\ no,\ delayed,\ miss)$

## Inferenza

Nella lezione precedente abbiamo visto l'inferenza tramite *entailment*. Ciò significa che possiamo dedurre nuove informazioni basandoci su informazioni già presenti. Possiamo anche inferire nuove informazioni basandoci sulle probabilità.

Questo non ci permette di dare per certo le nuove conoscenze, ma ci consente di capire le distribuzioni di probabilità di alcuni valori.

### Componenti dell'inferenza nelle reti Bayesiane

- Query **X**
  - la variabile per cui vogliamo computare la distribuzione di probabilità

- Variabili evidenza **E**
  - una o più variabili che sono state osservate per l'evento **e**

- Variabili nascoste **Y**
  - una o più variabili che non sono la query e che non sono state osservate

- **Goal**
  - Calcolare $P(X \mid e)$

#### Esempio 1

Esempio seguendo le componenti dell'inferenza

- Assumiamo `Train` come query
- Possiamo osservare che c'è una pioggia leggera (`Rain = light`), e questo ci aiuta a computare la probabilità che il treno è in ritardo
- Stando alla stazione possiamo osservare che piove, ma non sappiamo se c'è manutenzione sui binari fuori dalla nostra vista. La variabile nascosta, qui, è `Manutenzione`
- Calcoliamo la distribuzione di probabilità della variabile `Treno`, basata sul'evidenza **e** che sappiamo essere pioggia leggera

#### Esempio 2

Voglio calcolare la distribuzione di probabilità della variabile `Appointment` date le prove che c'è pioggia leggera e nessuna manutenzione. Quindi:

- `Rain = light`
- `Maintenance = no`
- `Train` è una variabile nascosta
- $P(Appointment \mid light,\ no)=?$

Dalla sezione sulla Probabilità Congiunta, sappiamo che possiamo esprimere i possibili valori della variabile aleatoria `Appointment` come proporzione, sapendo che:

$P(Appointment \mid light,\ no)=αP(Appointment,\ light,\ no)$

Come calcoliamo la distribuzione di probabilità di `Appointment` se il suo unico genitore è `Train` e non `Rain` o `Maintenance`?

Usiamo la Marginalizzazione. Ottengo:

$P(Appointment,\ light,\ no)=α[P(Appointment,\ light,\ no,\ delayed) + P(Appointment,\ light,\ no,\ on\ time)]$

## Inferenza tramite enumerazione

L'inferenza tramite enumerazione è il processo di trovare la distribuzione di probabilità di una variabile **X** date le prove osservate **e** e alcune variabili nascoste **Y**

È l'esplicitazione dei processi e delle componenti seguite nel paragrafo precedente.

Formula:



- **X** è la variabile query
- **e** è l'evidenza
- **y** è una delle variabili nascoste **Y**
- **α** normalizza il risultato, in modo che la somma dei membri del vettore risultante sia pari a 1

### Spiegazione formula

La distribuzione di probabilità di **X**, data **e**, è uguale alla distribuzione normalizzata di probabilità di **X** e **e**.

Per ottenere questa distribuzione, sommiamo la probabilità normalizzata di **X**, **e** e **y**, iterando nell'intervallo **Y**

### Codice di inferenza tramite enumerazione

Ci sono molte librerie che aiutano nei calcoli che includono inferenze e reti bayesiane, ma in questo caso utilizzeremo: `pomegranate`, che deve essere accompagnato da `pytorch` e/o `tensorflow`

- `Categorical` si riferisce a nodi senza genitori
- `ConditionalCategorical` è con i nodi con genitori, e si scrive come una lista di liste, quindi una matrice, come se stessi scrivendo la tabella vista sopra.

```python
# ruff: noqa: F403, F405

from pomegranate.distributions import Categorical, ConditionalCategorical
from pomegranate.bayesian_network import BayesianNetwork

# Define label mappings
rain_labels = ["none", "light", "heavy"]
maintenance_labels = ["yes", "no"]
train_labels = ["on time", "delayed"]
appointment_labels = ["attend", "miss"]

# Define distributions

# Rain node has no parents
rain = Categorical([[0.7, 0.2, 0.1]])  # P(rain)

# P(maintenance | rain)
maintenance = ConditionalCategorical([
    [[0.4, 0.6],  # rain = none
     [0.2, 0.8],  # rain = light
     [0.1, 0.9]]  # rain = heavy
])

# P(train | rain, maintenance)
train = ConditionalCategorical([
    [
        [[0.8, 0.2], [0.9, 0.1]],  # rain = none: maintenance = yes/no
        [[0.6, 0.4], [0.7, 0.3]],  # rain = light
        [[0.4, 0.6], [0.5, 0.5]]   # rain = heavy
    ]
])

# P(appointment | train)
appointment = ConditionalCategorical([
    [[0.9, 0.1],  # train = on time
     [0.6, 0.4]]  # train = delayed
])
```

Poi creiamo il modello, aggiungendo tutti i nodi e descrivendo quale nodo è il genitore di quale altro nodo, aggiungendo le `edge` tra i nodi.

Ricorda che la rete Bayesiana è un grafo direzionato, quindi nelle tuple delle `edge`, il primo nodo è sempre il genitore.

```python
# Build the Bayesian Network
model = BayesianNetwork(
    distributions=[rain, maintenance, train, appointment],
    edges=[(rain, maintenance), (rain, train), (maintenance, train), (train, appointment)]
)
```

Ora, per chiedere quale è la probabilità di un certo evento facciamo "partire" il modello con i valori a cui siamo interessati.

In questo esempio, vogliamo chiedere la probabilità del caso: niente pioggia, niente manutenzione, il treno è in orario e partecipiamo all'appuntamento.

```python
# Calculate probability for a given observation
# The numbers are the index value for each node
probability = model.probability([[0, 1, 0, 0]])

print(probability)
```

Altrimenti possiamo usare il programma per calcolare le distribuzioni di probabilità per tutte le variabili data un'evidenza osservata.

Nel caso seguente, sappiamo che il treno è in ritardo. Sapendo questo, calcoliamo e visualizziamo la distribuzione delle probabilità delle variabili `Rain`, `Maintenance`, e `Appontment`.

```python
import torch
from torch.masked import MaskedTensor

# Evidence: train = "delayed" (index 1), others unknown
evidence = torch.tensor([[ -1, -1, 1, -1 ]])
mask = evidence != -1
masked_input = MaskedTensor(evidence, mask)

# Run inference
predictions = model.predict_proba(masked_input)[0]

# Label mappings
label_maps = [rain_labels, maintenance_labels, train_labels, appointment_labels]
node_names = ["Rain", "Maintenance", "Train", "Appointment"]

# Print predictions
for i, prediction in enumerate(predictions):
    print(f"{node_names[i]}:")
    
    # If it's a tensor (distribution)
    if isinstance(prediction, torch.Tensor):
        for j, prob in enumerate(prediction.tolist()):
            print(f"  {label_maps[i][j]}: {prob:.4f}")
    
    # If it's a known value (int, NumPy int, etc.)
    else:
        try:
            index = int(prediction)
            for j, label in enumerate(label_maps[i]):
                prob = 1.0 if j == index else 0.0
                print(f"  {label}: {prob:.4f} (inferred)")
        except Exception:
            print(f"  Could not interpret prediction: {prediction} ({type(prediction)})")


print(predictions)
```

# Sampling (Campionamento)

Il Sampling è un modo per fare inferenza approssimata in una rete Bayesiana.

Prendo una "sample" (campione) di tutte le variabili del grafo, estraendo un valore per ogni nodo in base alla distribuzione di probabilità.

Questo genererà una distribuzione approssimativa dei valori ottenuti lanciando un dado: da un lato, è improbabile che otteniamo il risultato in cui ogni valore ha esattamente una probabilità di 1/6 di verificarsi (che è la probabilità teorica), ma otterremo un valore che si avvicina a essa.

##### Note: Rejection Sampling

Data un'evidenza **e**, se calcolo la probabilità condizionata assumendo che `Train` sia la mia **e**, questo calcolo prende il nome di **Rejection Sampling**

### Codice esempio del campionamento

```python
from model import model  # Assumes model is a BayesianNetwork built with pomegranate 1.0+
from collections import Counter # Will be used in next cell

def generate_sample():
    # Mapping of random variable name to sample generated
    sample = {}

    # Sample from the model (returns list of indices)
    sample_indices = model.sample(n=1)[0]

    # Variable names in the same order as defined in the model
    variable_names = ["rain", "maintenance", "train", "appointment"]

    # Mapping from index to human-readable label
    label_maps = {
        "rain": ["none", "light", "heavy"],
        "maintenance": ["yes", "no"],
        "train": ["on time", "delayed"],
        "appointment": ["attend", "miss"]
    }

    # We do not import `pomegranate` here because we are only using the pre-built model
    # and calling its `.sample()` method. No new distributions or types are being created.

    # Loop over all variables and convert sampled indices to labels
    for name, index in zip(variable_names, sample_indices):
        sample[name] = label_maps[name][index]

    # Return generated sample
    return sample
```

Ora, per calcolare $P(Appointment |\ Train = delayed)$, che è la distribuzione di probabilità della variabile `Appointment` dato `Train = delayed`, facciamo:

```python
# Rejection sampling
# Compute distribution of Appointment given that train is delayed
N = 10000
data = []

# Repeat sampling 10,000 times
for i in range(N):

    # Generate a sample based on the function that we defined earlier
    sample = generate_sample()

    # If, in this sample, the variable of Train has the value delayed, save the sample.
    # Since we are interested in the probability distribution of Appointment given that the train is delayed,
    # we discard the samples where the train was on time.
    if sample["train"] == "delayed":
        data.append(sample["appointment"])

# Count how many times each value of the variable appeared.
# We can later normalize by dividing the results by the total number of saved samples
# to get the approximate probabilities of the variable that add up to 1.
print(Counter(data))
```

Ricorda che il risultato non è normalizzato.

Ricorda che questo codice rappresenta Rejection Sampling.

Tuttavia se l'evento è estremamente improbabile, butteremmo via un sacco di campioni.

Se calcolo $P(X \mid Y)$ e Y accade 1 volta su 1000, butterò via in media 999 campioni, che è parecchio inefficiente, in quanto ho speso tempo a calcolarli.

# Modelli di Markov

o Processi di Markov

Finora si sono affrontati casi in cui le variabili erano immutabili. Tuttavia, può capitare, nella realtà, che cambino nel tempo.

Per valutare questi casi, definiamo una variabile $X$ e un generico evento.

- Nell'evento attuale: $X=X_t$ *oppure* $X=X(t)_n$
- Nell'evento successivo: $X=X_{t+1}$ *oppure* $X=X(t)_{n+1}$

## Assunzione di Markov

Lo stato corrente dipende solo su un numero finito di stati precedenti

## Catena di Markov

**Definizione:**

Una sequenza di variabili aleatorie la cui distribuzione, per ogni variabile, segue l'assunzione di Markov.
___

Ciò vuol dire che ogni evento nella catena si verifica basandosi sulla probabilità dell'evento precedente.

Per iniziare a costruire una catena di Markov, abbiamo bisogno di un **modello di transizione** che specifica le distribuzioni di probabilità del prossimo evento basandosi sui possibili valori dell'evento attuale

### Esempio: Meteo



In quest'esempio, la probabilità che domani sia soleggiato è basata sul fatto che, oggi, soleggiato è 0.8. Questo è ragionevole perché è molto più probabile che un giorno soleggiato ne segua uno soleggiato.

Tuttavia, se oggi piove la probabilità di pioggia, domani, è 0.7, poiché è molto più probabile che giorni piovosi si susseguano.

Usando questo modello di transizione, è possibile campionare una *Catena di Markov*.

Iniziamo da un giorno, sia esso soleggiato o piovoso, poi campioniamo il giorno successivo basandoci sulla probabilità che sia soleggiato o piovoso dato il tempo di oggi.

Poi, condizioniamo la probabilità di dopodomani basandoci su domani, e così via, ottenendo una Catena di Markov:



#### Esempio in codice

```python
import numpy as np
from pomegranate.distributions import Categorical, ConditionalCategorical
from pomegranate.markov_chain import MarkovChain

# Define state labels
states = ["sun", "rain"]
index_to_state = {0: "sun", 1: "rain"}

# Initial distribution: shape (1, 2)
start = Categorical(np.array([[0.5, 0.5]]))

# Transition matrix: shape (1, 2, 2)
transitions = np.array([[
    [0.8, 0.2],  # from sun
    [0.3, 0.7]   # from rain
]])

# Conditional distribution
transition_dist = ConditionalCategorical(probs=transitions)

# Create Markov Chain
model = MarkovChain([start, transition_dist])

# Sample 50 steps
samples = model.sample(50).flatten()

# Decode indices back to state names
decoded = [index_to_state[int(i.item())] for i in samples]
print(decoded)
```

## Modelli di Markov Nascosti

**Definizione:**

Un Modello di Markov Nascosto è un Modello di Markov per un sistema con stati nascosti che generano eventi osservati
___

Questo vuol dire che, a volte l'IA ha alcune misure del mondo ma nessun accesso allo stato preciso del mondo. In questi casi, lo stato del mondo è chiamato **stato nascosto** e qualsiasi informazione a cui l'IA ha accesso è detta **osservazione**

#### Esempio

| Hidden State | Observation |
| --- | --- |
| Posizione del robot | Dati dai sensori del robot |
| Parole pronunciate | Forme d'onda audio |
| Intrattenimento dell'utente | Website o App analytics |

### Sensor Model

o **Emission model**

Si occupano di tradurre gli Hidden States (stati nascosti), cioè lo stato dell'ambiente circostante, a partire dalle Observations (osservazioni) ottenute attraverso recettori di informazioni.

#### Esempio Sensor model

L'IA vuole inferire il meteo (stato nascosto) ma ha accesso solo a una telecamera interna che registra quante persone sono entrate con un ombrello.

L'immagine di seguito è il *sensor model* che rappresenta le probabilità.



In questo modello, se è soleggiato è più probabile che le persone non portino ombrelli nell'edificio.

Se è piovoso, allora è molto probabile che ne portino uno.

Usando queste osservazioni, possiamo predire che tempo fa all'esterno.


### Sensor Markov Assumption

(Assunzione sensoriale di Markov)

L'assunzione che la variabile evidenza dipende solamente dallo stato corrispondente.
___

Ciò non è necessariamente una riflessione della verità perché, ad esempio, persone più coscienziose o che odiano la pioggia possono portare un ombrello con sé ovunque anche se è soleggiato, e se conoscessimo le personalità di tutti avremmo più dati da aggiungere al modello.

Tuttavia, l'assunzione sensoriale di Markov ignora questi dati e presume che solo lo stato nascosto influenzi l'osservazione.

### Rappresentazione del Modello di Markov Nascosto

Posso rappresentarlo come una catena di Markov a due livelli.
- Il livello superiore, la variabile $X$, rappresenta lo stato nascosto.
- Il livello inferiore, la variabile $W$, rappresenta l'evidenza: l'osservazione che abbiamo



Basandoci sul Modello di Markov Nascosto, possiamo svolgere diversi compiti:

- **Filtraggio**
  - Date le osservazioni dall'inizio fino ad ora, calcola la distribuzione di probabilità per lo stato corrente.
  - Ad esempio, data l'informazione di quando le persone portano gli ombrelli dall'inizio del tempo fino ad oggi, generiamo una distribuzione di probabilità sul fatto che oggi piova o meno.

- **Predizione**
  - Date le osservazioni dall'inizio fino ad ora, calcola la distribuzione di probabilità per uno stato futuro.

- **Smoothing** (levigatura)
  - Date le osservazioni dall'inizio fino ad ora, calcola la distribuzione di probabilità per uno stato passato.
  - Ad esempio, calcolare la probabilità della pioggia per ieri dato che le persone hanno portato l'ombrello oggi.

- **Spiegazione più probabile**
  - Date le osservazioni dall'inizio fino ad ora, calcola la sequenza più probabile di eventi.

L'azione *Spiegazione più probabile* può essere usata in processi come riconoscimento vocale dove, basandoci su più forme d'onda, l'IA può inferire la sequenza di parole o sillabe più probabile che hanno generato quella forma d'onda.

#### Esempio in codice

```python
import numpy as np
from pomegranate.distributions import Categorical
from pomegranate.hmm import DenseHMM

# --- 1. Definizione del modello HMM (con dtype=np.float32 per tutti i float) ---

# Stati e osservazioni
states = ["sun", "rain"]
observations_mapping = ["umbrella", "no umbrella"]
obs_to_index = {obs: i for i, obs in enumerate(observations_mapping)}
index_to_obs = {i: obs for obs, i in obs_to_index.items()}
index_to_state = {i: state for i, state in enumerate(states)} # Mappa indici a nomi stati

# Distribuzioni di emissione (P(obs | state))
# IMPORTANTE: Specificare dtype=np.float32
sun = Categorical(np.array([[0.2, 0.8]], dtype=np.float32))
rain = Categorical(np.array([[0.9, 0.1]], dtype=np.float32))

emissions = [sun, rain]

# Matrice di transizione
# IMPORTANTE: Specificare dtype=np.float32
transitions = np.array([
    [0.8, 0.2],  # sun → sun, sun → rain
    [0.3, 0.7]   # rain → sun, rain → rain
], dtype=np.float32)

# Probabilità iniziali
# IMPORTANTE: Specificare dtype=np.float32
starts = np.array([0.5, 0.5], dtype=np.float32)

# Probabilità di uscita (facoltativa, può essere zero)
# IMPORTANTE: Specificare dtype=np.float32
ends = np.array([0.0, 0.0], dtype=np.float32)

# Costruzione del modello HMM
model = DenseHMM(
    distributions=emissions,
    edges=transitions,
    starts=starts,
    ends=ends
)
```

Nota che il modello ha sia il sensor model che il transition model. Abbiamo bisogno di entrambi per l'Hidden Markov Model.

Nel codice seguente, vediamo una sequenza di osservazioni nel caso le persone abbiano portato un ombrello nell'edificio o meno, e basandoci su questa sequenza, avviaremo il modello, che genererà e mostrerà la spiegazione più probabile (cioè la sequenza meteo più probabile che ha portato a questo pattern di osservazioni)

```python
# --- 2. Preparazione dei dati osservati e previsione ---

# Dati osservati come stringhe
observed_data_str = [
    "umbrella",
    "umbrella",
    "no umbrella",
    "umbrella",
    "umbrella",
    "umbrella",
    "umbrella",
    "no umbrella",
    "no umbrella"
]

# Converti le osservazioni da stringhe a indici numerici
# Questo array contiene interi, quindi non necessita di dtype=np.float32 qui
observed_data_indices = np.array([obs_to_index[obs] for obs in observed_data_str])

# IMPORTANTE: Aggiungi le dimensioni mancanti per ottenere la forma (n_sequences, n_timesteps, n_features)
# Una singola sequenza, lunghezza della sequenza, 1 feature per timestep
observed_data_3d = observed_data_indices[np.newaxis, :, np.newaxis]

# Esegui la previsione degli stati sottostanti
predictions = model.predict(observed_data_3d)

# --- 3. Stampa dei risultati ---

print("Sequenza di osservazioni:")
print(observed_data_str)
print("\nStati previsti:")

# Dato che predict restituisce un array 2D (anche per una singola sequenza),
# dobbiamo accedere al primo (e unico) elemento della prima dimensione.
# IMPORTANTE: Usa .item() per estrarre l'intero dal tensore/array NumPy.
for prediction_tensor in predictions[0]: # prediction_tensor sarà un tensore/array NumPy con un singolo elemento
    prediction_index = prediction_tensor.item() # Estrai il valore int
    print(index_to_state[prediction_index])
```
