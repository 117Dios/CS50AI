# Search

"Risolvere un labirinto", come ad esempio trovare la strada con un GPS

### Terminologia

**Agente / Agent**
: entità che percepisce l'ambiente e agisce su quello stesso ambiente

**Stato / State**
: configurazione dell'agente e del suo ambiente

- **Stato iniziale / Initial state**
: stato in cui l'agente inizia

**Azioni / Actions**
: scelte che si possono fare in uno stato. Le azioni restituiscono
- `ACTION(s)` restituisce il set di azioni che possono essere effettuate in uno stato `s`

**Modello di transizione / Transition model**
: una descrizione di quale stato risulta dall'esecuzione di una qualsiasi azione eseguibile in un qualunque stato

**Spazio degli stati / State space**
: set di tutti gli stati raggiungibili dallo stato iniziale da una qualsiasi sequenza di azioni
- Grafo: i cerchi sono gli stati, le frecce le azioni



**Goal Test**
: modo per l'IA di determinare se un certo stato è uno stato "goal" (destinazione o successo)
- Possono esserci multipli stati goal

**Path Cost**
: costo numerico associato a un percorso dato
- Utile per arrivare a uno stato goal con una certa efficienza



**Soluzione**
: sequenza di azioni che portano dallo stato iniziale allo stato goal

**Soluzione ottimale**
: soluzione/i che ha il path cost minore tra tutte le soluzioni

**Nodo**
: struttura dati che tiene traccia di:
- uno stato
- un genitore (nodo che ha generato questo nodo)
- un'azione (azione applicata al genitore per ottenere il nodo)
- un Path cost (da stato iniziale a nodo)

# Solving Search Problems

## Composizione della soluzione dei Search Problems

- **Stato iniziale**
: da dove iniziamo
- **Azioni**
: una o più azioni possibili nei vari stati
- **Modello di transizione**
: modo per definire, dato uno stato attuale e un'azione eseguita, in che stato finiremo 
- **Goal test**
: per sapere se abbiamo raggiungo il nostro goal
- **Path cost function**
: per definire quanto è efficiente il percorso seguito per raggiungere il goal

## Soluzione
- Inizia con una **frontiera** che contiene lo stato iniziale.
  - Frontiera: tutti gli stati raggiungibili dagli stati attuali. La prima volta contiene lo stato iniziale perché siamo "fuori" dal grafo
- Dopodiché, ripeti:
  1. Se la frontiera è vuota: `Nessuna soluzione`
  2. Rimuovi un nodo dalla frontiera (Lo stiamo "selezionando")
  3. Se il nodo contiene un goal state, restituisci la soluzione
  4. Altrimenti *Espandi il nodo*, cioè considera tutte le azioni eseguibili in quel nodo e gli stati che ne derivano. Poi, aggiungi quei nodi dalla frontiera

## Approccio revisionato

- Inizia con una frontiera che contiene lo stato iniziale
- Inizializza un `set` vuoto di nodi già esplorati
- Dopodiché, ripeti:
  1. Frontiera vuota: `nessuna soluzione`
  2. Rimuovi un nodo dalla frontiera
  3. Se il nodo contiene il goal state, restituisci la soluzione
  4. Aggiungi il nodo al `set` di nodi esplorati
  5. Espandi il nodo, aggiungi i nodi risultanti alla frontiera se non già nella frontiera o nel set di nodi esplorati

La frontiera è una struttura dati, è quindi molto importante decidere come è strutturata, come aggiungiamo e come rimuoviamo nodi da essa.
