````python
# Define the function that removes a node from the frontier and returns it.
def remove(self):
        # Terminate the search if the frontier is empty, because this means that there is no solution.
    if self.empty():
        raise Exception("empty frontier")
    else:
        # Save the oldest item on the list (which was the first one to be added)
        node = self.frontier[0]
        # Save all the items on the list besides the first one (i.e. removing the first node)
        self.frontier = self.frontier[1:]
        return node
````

# A* search

Algoritmo di ricerca che espande il nodo con il valore pi basso di $g(n)+h(n)$, con:
- $g(n)=$ costo per raggiungere il nodo
- $h(n)=$ costo stimato per il goal

Ci che faccio, quindi,  contare il numero di passi fatti e sommarli alla distanza di Manhattan di ogni nodo in cui arrivo. Osservo poi la somma dei nodi successivi in cui voglio muovermi.

Se questa somma aumenta rispetto al passo precedente, vado in un'altra direzione, dopodich:
* Se la somma lungo la nuova direzione  uguale a quella che avevo, il percorso viene scelto randomicamente tra tutti quelli con lo stesso valore.
* Se  minore, la seguo e reitero le operazioni fatte finora

Ricorda che il numero di passi  la distanza dallo start state, non quanti passi ho fatto finora.

Note:

A* usa un sacco di memoria

## Quando A*  ottimale?

A*  ottimale se valgono entrambe le seguenti affermazioni:

- $h(n)$  ammissibile
  - Cio, non sovrastima mai il vero costo
- $h(n)$  consistente
  - Per ogni nodo $n$ e successore $n'$ con costo di passo $c$, devo avere: $h(n) \leq h(n')+c$

# Minimax

Ottimo per giochi deterministici con 2 giocatori, a mosse alterne. L'esempio generico che sar usato  tris (tic-tac-toe in inglese)

Minimax restituir +1 in caso di vittoria per una delle due parti, -1 in caso di vittoria della parte opposta e 0 in caso di pareggio.



## Algoritmo

Sceglieremo un MAX player e un MIN player
- MAX cercher di massimizzare il suo punteggio
  - In questo caso scegliamo X come MAX
- MIN cercher di minimizzare il suo punteggio
  - In questo caso scegliamo O come MIN

MAX sceglier, il pi possibile, di ottenere +1. Se non riesce, per, preferir ottenere un pareggio piuttosto che perdere.

Viceversa, MIN preferisce perdere ma, se non riesce, preferisce il pareggio.

## Come impostare il gioco

- $s_{0}: $ stato iniziale
- `PLAYER(s)`: restituisce quale giocatore deve agire nello stato `s`
- `ACTION(s)`: restituisce le azioni possibili nello stato `s`
- `RESULT(s,a)`: restituisce lo stato raggiunto dopo l'azione `a` effettuata nello stato `s`
- `TERMINAL(s)`: controlla se lo stato `s`  uno stato finale
- `UTILITY(s)`: restituisce il valore numerico finale dallo stato finale `s`

## Formalizzazione Minimax

Fatta con pseudocodice

- Dato uno stato `s`
  - `MAX` sceglie l'azione `a` in `ACTION(s)` che produce il valore pi alto di `MIN-VALUE(RESULT(s,a))`
  - `MIN` sceglie l'azione `a` in `ACTION(s)` che produce il valore pi basso di `MAX-VALUE(RESULT(s,a))`

`MAX` e `MIN` stanno quindi cercando di capire cosa vuole fare l'altro.

Una volta capito come implementare `MIN-VALUE(RESULT(s,a))` e `MAX-VALUE(RESULT(s,a))` ho effettivamente un'implementazione completa di Minimax

### MAX-VALUE

Innanzitutto controllo se sono nello stato finale, cos in caso UTILITY(s) mi restituisce subito il punteggio.

Altrimenti, cerco di massimizzare il valore. Inizio dichiarando una variabile v e la pongo al valore pi piccolo possibile, perch voglio cercare sempre di far meglio di v, e impostandola cos bassa far sempre meglio di -.

Dopodich, nel for, v sar uguale al massimo tra v e il meglio che MIN riuscir a fare in quello stato, data un'azione possibile in quello stato.

```
function MAX-VALUE(state):
  if TERMINAL(state):
    return UTILITY(state)

  for action in ACTIONS(state):
    v = MAX(v, MIN-VALUE(RESULT(state, action)))

  return v
```

## Depth-Limited Minimax

A volte, per, gli stati possibili sono troppi da valutare. Ecco dove entra in gioco quest'ottimizzazione.

Normalmente Minimax valuta TUTTI gli stati. Quest'ottimizzazione, invece, limita quanti stati guarda in avanti, e ignora i successivi.

Se si arriva a quel limite ma il gioco non  ancora finito, che facciamo? Chiamiamo un'evaluation function.

**evaluation function**
: funzione che stima la UTILITY prevista del gioco dato un certo stato.
