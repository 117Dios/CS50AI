# A* search

Algoritmo di ricerca che espande il nodo con il valore più basso di $g(n)+h(n)$, con:
- $g(n)=$ costo per raggiungere il nodo
- $h(n)=$ costo stimato per il goal

Ciò che faccio, quindi, è contare il numero di passi fatti e sommarli alla distanza di Manhattan di ogni nodo in cui arrivo. Osservo poi la somma dei nodi successivi in cui voglio muovermi.

Se questa somma aumenta rispetto al passo precedente, vado in un'altra direzione, dopodiché:
* Se la somma lungo la nuova direzione è uguale a quella che avevo, il percorso viene scelto randomicamente tra tutti quelli con lo stesso valore.
* Se è minore, la seguo e reitero le operazioni fatte finora

Ricorda che il numero di passi è la distanza dallo start state, non quanti passi ho fatto finora.

Note:

A* usa un sacco di memoria

## Quando A* è ottimale?

A* è ottimale se valgono entrambe le seguenti affermazioni:

- $h(n)$ è ammissibile
  - Cioè, non sovrastima mai il vero costo
- $h(n)$ è consistente
  - Per ogni nodo $n$ e successore $n'$ con costo di passo $c$, devo avere: $h(n) \leq h(n')+c$

# Minimax

Ottimo per giochi deterministici con 2 giocatori, a mosse alterne. L'esempio generico che sarà usato è tris (tic-tac-toe in inglese)

Minimax restituirà +1 in caso di vittoria per una delle due parti, -1 in caso di vittoria della parte opposta e 0 in caso di pareggio.



## Algoritmo

Sceglieremo un MAX player e un MIN player
- MAX cercherà di massimizzare il suo punteggio
  - In questo caso scegliamo X come MAX
- MIN cercherà di minimizzare il suo punteggio
  - In questo caso scegliamo O come MIN

MAX sceglierà, il più possibile, di ottenere +1. Se non riesce, però, preferirà ottenere un pareggio piuttosto che perdere.

Viceversa, MIN preferisce perdere ma, se non riesce, preferisce il pareggio.

## Come impostare il gioco

- $s_{0}: $ stato iniziale
- `PLAYER(s)`: restituisce quale giocatore deve agire nello stato `s`
- `ACTION(s)`: restituisce le azioni possibili nello stato `s`
- `RESULT(s,a)`: restituisce lo stato raggiunto dopo l'azione `a` effettuata nello stato `s`
- `TERMINAL(s)`: controlla se lo stato `s` è uno stato finale
- `UTILITY(s)`: restituisce il valore numerico finale dallo stato finale `s`

## Formalizzazione Minimax

Fatta con pseudocodice

- Dato uno stato `s`
  - `MAX` sceglie l'azione `a` in `ACTION(s)` che produce il valore più alto di `MIN-VALUE(RESULT(s,a))`
  - `MIN` sceglie l'azione `a` in `ACTION(s)` che produce il valore più basso di `MAX-VALUE(RESULT(s,a))`

`MAX` e `MIN` stanno quindi cercando di capire cosa vuole fare l'altro.

Una volta capito come implementare `MIN-VALUE(RESULT(s,a))` e `MAX-VALUE(RESULT(s,a))` ho effettivamente un'implementazione completa di Minimax

### MAX-VALUE

Innanzitutto controllo se sono nello stato finale, così in caso UTILITY(s) mi restituisce subito il punteggio.

Altrimenti, cerco di massimizzare il valore. Inizio dichiarando una variabile v e la pongo al valore più piccolo possibile, perché voglio cercare sempre di far meglio di v, e impostandola così bassa farò sempre meglio di -∞.

Dopodiché, nel for, v sarà uguale al massimo tra v e il meglio che MIN riuscirà a fare in quello stato, data un'azione possibile in quello stato.

```
function MAX-VALUE(state):
  if TERMINAL(state):
    return UTILITY(state)

  v = -∞

  for action in ACTIONS(state):
    v = MAX(v, MIN-VALUE(RESULT(state, action)))

  return v
```

### MIN-VALUE

Discorso analogo e opposto per MIN-VALUE

Innanzitutto controllo se sono nello stato finale, così in caso UTILITY(s) mi restituisce subito il punteggio.

Altrimenti, cerco di minimizzare il valore. Inizio dichiarando una variabile v e la pongo al valore più grande possibile, perché voglio cercare sempre di far peggio di v, e impostandola così alta farò sempre meglio di ∞.

Dopodiché, nel for, v sarà uguale al minimo tra v e il peggio che MAX riuscirà a fare in quello stato, data un'azione possibile in quello stato.

```
function MIN-VALUE(state):
  if TERMINAL(state):
    return UTILITY(state)

  v = ∞

  for action in ACTIONS(state):
    v = MIN(v, MAX-VALUE(RESULT(state, action)))

  return v
```

Valuto il grafico come giocatore MAX. Inizialmente, sono disponibili solo le mie mosse iniziali, cioè quelle in basso.

Al momento, mi trovo al triangolo solitario in alto. Non ho un valore, voglio dedurlo dai triangoli rossi, che a loro volta dedurranno i loro valori dai miei input.

2. I miei valori che ho inserito lì sono 4, 8 e 5
3. MIN sicuramente dirà che diventerà 4
4. Esploro il triangolo in mezzo
5. Ho inserito 9, 3 e 7. MIN sceglierà sicuramente 3
6. Valuto il triangolo rosso a destra
7. Ho inserito 2, 4 e 6
8. MIN sceglierà sicuramente 2
9. MIN avrà quindi inserito 4, 3 e 2
10. Sceglierò quindi 4.

#### Ripeto il ragionamento



1. Valuto il triangolo a sinistra. Il ragionamento è uguale
2. Valuto il triangolo centrale
Non appena vedo il 3, so che è minore di 4, quindi al massimo il valore di quel triangolo è 3
3. Posso ignorare il mio terzo input nei 3 triangoli centrali
4. Valuto il triangolo rosso a destra
5. Trovo un 2. Il triangolo rosso sarà al massimo 2, quindi posso ignorare gli altri miei input
6. Assumo il valore di 4 in cima

Ignorare nodi e quindi non valutarli "mai più" equivale a tagliarli, cioè *pruning*, mentre tengo conto dei due valori a confronto quando valuto, *alpha* e *beta*

## Depth-Limited Minimax

A volte, però, gli stati possibili sono troppi da valutare. Ecco dove entra in gioco quest'ottimizzazione.

Normalmente Minimax valuta TUTTI gli stati. Quest'ottimizzazione, invece, limita quanti stati guarda in avanti, e ignora i successivi.

Se si arriva a quel limite ma il gioco non è ancora finito, che facciamo? Chiamiamo un'evaluation function.

**evaluation function**
: funzione che stima la UTILITY prevista del gioco dato un certo stato.
