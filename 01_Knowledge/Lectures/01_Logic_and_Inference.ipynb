## Concetti

- **knowledge-based agents**
: agenti che ragionano operando su rappresentazioni interne di informazioni. Sono generalmente le IA

- **sentence (o frase)**
: un'asserzione sul mondo in ciò che chiameremo "linguaggio rappresentativo di conoscenza"

Una frase è come un'IA conserva l'informazione e la usa per inferire nuove informazioni

## Propositional logic

Logica basata su proposizioni: affermazioni sul mondo che possono essere vere o false.

### Simboli di proposizione

Ogni simbolo rappresenta una proposizione differente

- $P$
- $Q$
- $R$

### Connettori logici

Sotto saranno rappresentati i 5 più importanti, ma ce ne sono altri

#### Not: $¬$

¬ inverte la tabella della verità di una proposizione

Esempio e tabella della verità:

| Simbolo | Proposizione |
| --- | --- |
| P | Oggi piove|
| ¬P | Oggi non piove|

#### And: $∧$

∧ connette due proposizioni diverse. Prendendo P e Q generiche e connettendole con ∧, il risultato P∧Q sarà vero solo se P e Q saranno vere se prese singolarmente.

Tabella della verità:

| Simboli | Proposizione |
|---| --- |
| P | False |
| Q | False |
| P∧Q | False|
| --- | --- |
| P | False |
| Q | True |
| P∧Q | False|
| --- | --- |
| P | True |
| Q | False |
| P∧Q | False|
| --- | --- |
| P | True |
| Q | True |
| P∧Q | True|

#### Or: $∨$

∨ risulta vero finché uno dei due argomenti è vero. Quindi, affinché P∨Q sia vero, basta che P o Q sia vero.

È giusto precisare che esistono:
- Or inclusivo (Or, simbolo: ∨)
- Or esclusivo (XOR, simbolo: ⊕)

In XOR, P∨Q è falso se P∧Q è vero. Quindi in XOR solo P o Q devono essere veri, ma non entrambi.

Quando scriviamo Or generalmente si intende Or inclusivo, che rappresentiamo nella tabella sotto.

Tabella della verità:

| Simboli | Proposizione |
|---| --- |
| P | False |
| Q | False |
| P∨Q | False|
| --- | --- |
| P | False |
| Q | True |
| P∨Q | True|
| --- | --- |
| P | True |
| Q | False |
| P∨Q | True|
| --- | --- |
| P | True |
| Q | True |
| P∨Q | True|

#### Implication: $→$ (Implicazione logica)

Rappresenta la struttura "Se P, allora Q.", quindi P implica Q.

Quando P→Q, P è chiamato **antecedente** e Q **conseguente**

Esempio:
- P: Sta piovendo
- Q: Sono in casa
- P → Q: Se sta piovendo, allora sto in casa

Tabella della verità:

- Quando l'*antecedente* è **vero** e il *conseguente* è **vero**, l'implicazione è **vera**.
- Quando l'*antecedente* è **vero** e il *conseguente* è **falso**, l'implicazione è **falsa**.
- Quando l'*antecedente* è **falso**, l'implicazione è automaticamente **vera** a prescindere dal *conseguente*
  - Quest'implicazione è chiamata *trivialmente vera*. Cioè, è formalmente vera ma non ha contenuto informativo
  - Questo perché, con antecedente falso, è impossibile verificare che implichi il conseguente. Data l'affermazione "Se non sta piovendo, allora sto in casa", risulta vera a prescindere che io sia fuori o dentro casa. Questo perché "Se sta piovendo, allora sto in casa" non ci dice nulla riguardo a cosa accade se **non** piove. 

Tabella:

| Simboli | Proposizione|
| --- | --- |
| P | False |
| Q | False
| P → Q | True |
| --- | --- |
| P | False |
| Q | True
| P → Q | True |
| --- | --- |
| P | True |
| Q | False
| P → Q | False |
| --- | --- |
| P | True |
| Q | True
| P → Q | True |

#### Biconditional: $↔$ (Coimplicazione)

↔ indica che l'implicazione va in ambo le direzioni. Si può leggere come "Se e solo se".

P ↔ Q equivale a P → Q e Q → P contemporaneamente.

Esempio:

- P: "Sta piovendo"
- Q: "Sto in casa"
- P ↔ Q: "Se sta piovendo, allora sto in casa" e "Se sto in casa, allora sta piovendo"

Possiamo quindi inferire più informazioni da una semplice implicazione. Quindi se P è falso, Q è falso, e viceversa.

Tabella della verità:

| Simboli | Proposizione|
| --- | --- |
| P | False |
| Q | False
| P ↔ Q | True |
| --- | --- |
| P | False |
| Q | True
| P ↔ Q | False |
| --- | --- |
| P | True |
| Q | False
| P ↔ Q | False |
| --- | --- |
| P | True |
| Q | True
| P ↔ Q | True |

## Modello

Assegnazione di un valore di verità a ogni simbolo di proposizione. Cioè, una "parola possibile". Possiamo pensare a un modello come a un "mondo" in cui quella cosa è vera.

Esempio:
- P: Sta piovendo
- Q: Oggi è martedì
- Modello: {P = True, Q = False}

Ci sono comunque modelli in cui magari le proposizioni sono entrambe vere, entrambe false o invertite.

Infatti, per $n$ variabili propositive binarie (Vere o False), esistono $2^{n}$ modelli.

## Knowledge base (KB)

Un set di sentences (frasi) conosciute dal knowledge-based agent (IA)

La conoscenza data all'IA attravero le frasi di logica propositiva può essere usata per fare nuove inferenze sul mondo

## Entailment: $⊨$ (Conseguenza logica)

$α ⊨ β$ è un'implicazione vincolante. L'implicazione logica già vista è più debole rispetto all'entailment.

α entails β
- In ogni modello in cui la frase α è vera, la frase β è anch'essa vera

Invece, in α → β, β potrebbe essere falsa

Esempio:

Se α = "Oggi è un martedì di Gennaio" e β = "È Gennaio", allora sappiamo che α ⊨ β.
- Se è un martedì di gennaio, sappiamo che è Gennaio.

## Inference (Inferenza)

L'inferenza è il processo di derivazione di nuove frasi a partire da frasi vecchie. Ci sono più modi per derivare nuova conoscenza basandosi su conoscenza pre-esistente

Esempio:

- P: Oggi è martedì
- Q: Sta piovendo
- R: Harry andrà a correre

**KB**:  $(P ∧ ¬Q) → R$

Questo vuol dire che P (P=True) e not Q (Q=Falso) implicano R. PEr capire se R è vero o falso devo chiedermi: KB ⊨ R?

Quindi abbiamo bisogno di un algoritmo che ci aiuti a rispondere alla generica domanda: possiamo verificare che $KB ⊨ α$?

Ci sono vari algoritmi che lo fanno, uno dei quali è **Model Checking**

Quello che fa Model Checking è:

- Per determinare se $KB ⊨ α$ (Quindi, rispondere alla domanda: "Possiamo concludere che $α$ è vera basandoci sulla nostra knowledge base?")
  - Enumerare tutti i possibili modelli
  - Se in ogni modello dove $KB$ è vero, anche $α$ è vero, allora $KB$ entails $α$ $(KB ⊨ α)$
  - Altrimenti, $KB$ does not entail $α$

Quindi se in ogni "mondo" la nostra conoscenza è vera, lo sarà anche α.

### Model Checking

Iniziamo enumerando tutti i modelli.

A destra dell'inferenza indico P e ¬Q per dire P = True e ¬Q = True (Q = False)



Dopodiché, data la nostra Knowledge Base, controlliamo se i modelli sono veri.

Innanzitutto, nella nostra KB, sappiamo che P = True, quindi possiamo affermare che KB = False in tutti i modelli in cui P = False



Poi, sappiamo che nella nostra KB, Q è falso. Quindi, possiamo dire che KB = False in tutti i modelli in cui Q = True

Infine, ci rimarranno due modelli. In entrambi, P = True e Q = False. In uno dei due R = True e nell'altro R = False.

Data la nostra $KB: (P ∧ ¬Q) → R$, sappiamo che quando P = True e Q = False allora R = True.

Quindi, la nostra KB è falsa nel caso R sia falsa, mentre KB è vera quando R è vera.

C'è un solo modello in cui R = True, e lì la nostra KB è vera. Per la definizione di "entailmente", se R = True in tutti i modelli dove KB = True, allora KB ⊨ R



### Codice rappresentativo di Inference

La libreria [logic.py](logic.py) è stata creata dal Lecturer e non ci interessa per ora.

Il seguente snippet proviene da [harry.py](harry.py)

```python
# ruff: noqa: F403, F405 (soppressione dei warning per wildstar import)

from logic import *

# Create new classes, each having a name, or a symbol, representing each proposition.
rain = Symbol("rain")  # It is raining.
hagrid = Symbol("hagrid")  # Harry visited Hagrid
dumbledore = Symbol("dumbledore")  # Harry visited Dumbledore

# Save sentences into the KB
knowledge = And(  # Starting from the "And" logical connective, becasue each proposition represents knowledge that we know to be true.

    Implication(Not(rain), hagrid),  # ¬(It is raining) → (Harry visited Hagrid)

    Or(hagrid, dumbledore),  # (Harry visited Hagrid) ∨ (Harry visited Dumbledore).

    Not(And(hagrid, dumbledore)),  # ¬(Harry visited Hagrid ∧ Harry visited Dumbledore) i.e. Harry did not visit both Hagrid and Dumbledore.

    dumbledore  # Harry visited Dumbledore. Note that while previous propositions contained multiple symbols with connectors, this is a proposition consisting of one symbol. This means that we take as a fact that, in this KB, Harry visited Dumbledore.
    )

# Model check algorith to infer if it is raining based on current knowledge
model_check(knowledge,rain)

print(knowledge.formula())
print(f"Is it raining? {model_check(knowledge,rain)}")
```

#### Algoritmo Model Check

Per eseguire l'algoritmo di Model Checking, c'è bisogno delle seguenti informazioni:
- Knowledge Base, che sarà usata per le inferenze
- Una query, o la proposizione di cui ci interessa valutare l'entailing dalla KB
- Simboli: una lista di tutti i simboli (o proposizioni atomiche) usate
  - Nel nostro caso sono: `rain`, `hagrid` e `dumbledore`
- Modello: un'assegnazione dei valori Vero e Falso ai simboli

Di seguito l'algoritmo in codice:

```python
def model_check(knowledge, query):
    """Checks if knowledge base entails query."""

    def check_all(knowledge, query, symbols, model):

        # If model has an assignment for each symbol
        # (The logic below might be a little confusing: we start with a list of symbols.
        #   The function is recursive, and every time it calls itself it pops one symbol from the symbols list and generates models from it.
        #   Thus, when the symbols list is empty, we know that we finished generating models with every possible truth assignment of symbols.)
        if not symbols:

            # If knowledge base is true in model, then query must also be true
            if knowledge.evaluate(model):
                return query.evaluate(model)
            return True
        else:

            # Choose one of the remaining unused symbols
            remaining = symbols.copy()
            p = remaining.pop()

            # Create a model where the symbol is true
            model_true = model.copy()
            model_true[p] = True

            # Create a model where the symbol is false
            model_false = model.copy()
            model_false[p] = False

            # Ensure entailment holds in both models
            return(check_all(knowledge, query, remaining, model_true) and
                check_all(knowledge, query, remaining, model_false))

    # Get all symbols in both knowledge and query. Method model_check() starts here
    symbols = set.union(knowledge.symbols(), query.symbols())

    # Check that knowledge entails query
    return check_all(knowledge, query, symbols, dict())
```

#### Note

Da notare che siamo interessati solo ai modelli dove KB = True. Se KB = False, allora le condizioni che sappiamo essere vere non si stanno verificando in questi modelli, rendendoli irrilevanti

Inoltre, il modo in cui `check_all` funziona è ricorsivo:

1. Prende un simbolo
2. Crea due modelli
3. In uno dei due il simbolo è True, nell'altro è False
4. Si richiama, ora con due modelli che differiscono nell'assegnazione di verità di un simbolo

La funzione continuerà a fare ciò finché non sarà assegnato un valore di verità a ogni simbolo, lasciando la lista `symbols` vuota.

Una volta vuota (caso identificato dalla riga `if not symbols`), in tutte le istanze della funzione (in quanto ogni istanza contiene un modello diverso), la funzione controlla se KB è vera dato il modello.

Se KB è vera in questo modello, la funzione controlla se la query è vera, come descritto prima.
